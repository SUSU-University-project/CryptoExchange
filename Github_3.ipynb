#Генерация идей
###Крипто обменник
---
*	Децентрализованное приложение для обмена криптовалютами с сопряжением метамасок, получением данных смарт-контракта в реальном времени
*	Web приложение
*	Приложение будет распространяться бесплатно. Предполагается монетизация через встроенную рекламу.
*	Срок вывода продукта на рынок 4 месяца

###Мессенджер на блокчейне
---
* Блокчейн-мессенджеры зашифрованы, поэтому пользователь может быть уверен, что его информация и личные данные будут защищены, а также конфиденциальны.
*	Приложение для мобильных операционных систем iOS и Android, а так же web версия
*	Приложение будет распространяться бесплатно через официальные маркетплейсы платформ. Предполагается монетизация через рекламу, а так же внутренние покупки.
*	Срок вывода продукта на рынок 6 месяцев
###Приложение с программами тренировок
---
*	Приложение для составления индивидуальных тренировок и с демонстрацией выполнения упражнений
*	Web-сайт, приложение для мобильных операционных систем iOS и Android
*	Приложение будет распространяться по платной подписке
*	Срок вывода продукта на рынок 6 месяцев
